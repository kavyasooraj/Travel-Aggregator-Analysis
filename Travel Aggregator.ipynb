```python
2.#How many sessions have more than one booking ?

booked_sessions = sessions_df[sessions_df['booking_id'].notnull()]# Filter the dataframe  where 'booking_id' is not null
session_booking_counts = booked_sessions.groupby('session_id').size().reset_index(name='booking_count')# Group the sessions by 'session_id' 
multiple_bookings = session_booking_counts[session_booking_counts['booking_count'] > 1]
print(multiple_bookings)
```

```python
#3Which days of the week have the highest number of bookings? Also,draw a piechart to show the distribution for all days of the week?

import pandas as pd
import matplotlib.pyplot as plt

sessions_df['session_starting_time'] = pd.to_datetime(sessions_df['session_starting_time'], format='ISO8601')# ISO8601 used to handle timestamps properly
sessions_df['day_of_week'] = sessions_df['session_starting_time'].dt.day_name()
booked_sessions = sessions_df[sessions_df['booking_id'].notnull()]# Filter rows where there is a booking (i.e., 'booking_id' is not null)
bookings_by_day = booked_sessions['day_of_week'].value_counts()# Count the number of bookings for each day of the week
print(bookings_by_day)

plt.figure(figsize=(4, 4))
plt.pie(bookings_by_day, labels=bookings_by_day.index, autopct='%1.1f%%',  colors=plt.cm.Paired.colors)
plt.title('Bookings Distribution by Day of the Week')
plt.legend(loc="upper right", bbox_to_anchor=(1.9,1))
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#9.Plot the trends at a quarterly frequency for the number of bookings by each of the device types, that is, plot a 
#time series for each year and quarter showing the number of bookings performed by each device type
import pandas as pd
import matplotlib.pyplot as plt
bookings_df['booking_time'] = pd.to_datetime(bookings_df['booking_time'])
bookings_df.set_index('booking_time', inplace=True)
device_type_trends = bookings_df.groupby(['service_name', pd.Grouper(freq='Q'), 'device_type_used'])['booking_id'].count().unstack().fillna(0)
device_type_trends.groupby(level=0).plot(kind='line', subplots=True, figsize=(10, 20), legend=True, title='Device Type Trends by Service (Quarterly)')
plt.tight_layout()
plt.show()
```

```python
#10
import pandas as pd
import matplotlib.pyplot as plt
merged_df = pd.merge(sessions_df, bookings_df, on='booking_id', how='left')      #merge the dataframe 
total_searches = sessions_df['search_id'].nunique()                   # Unique searches from sessions
total_bookings = bookings_df['booking_id'].nunique()                 # Unique bookings
total_customers = len(bookings_df['customer_id'].unique())           # Unique customers
oBSR = total_bookings / total_searches                              # Calculate overall Booking Success Rate (oBSR)
print(f"Overall Booking Success Rate (oBSR): {oBSR:.2f}")
```

```python
merged_df['search_time'] = pd.to_datetime(merged_df['search_time'], utc=True, format='ISO8601')  # Convert 'search_time' to datetime
monthly_data = merged_df.groupby(merged_df['search_time'].dt.to_period('M')).agg(
    total_bookings=('booking_id', 'nunique'),  # Count unique bookings
    total_searches=('search_id', 'nunique')  # Count unique searches
).reset_index()
# Calculate oBSR for each month
monthly_data['oBSR'] = monthly_data['total_bookings'] / monthly_data['total_searches']
monthly_data['month'] = monthly_data['search_time'].dt.to_timestamp()

# Calculate weekly oBSR
weekly_data = merged_df.groupby(merged_df['search_time'].dt.to_period('W')).agg(
    total_bookings=('booking_id', 'nunique'),
    total_searches=('search_id', 'nunique')
).reset_index()

# Calculate oBSR for each week
weekly_data['oBSR'] = weekly_data['total_bookings'] / weekly_data['total_searches']
weekly_data['week'] = weekly_data['search_time'].dt.to_timestamp()

# Plotting Monthly oBSR
plt.figure(figsize=(12, 6))
plt.plot(monthly_data['month'], monthly_data['oBSR'], marker='*', label='Monthly oBSR')
plt.title('Monthly Overall Booking Success Rate (oBSR)')
plt.xlabel('Month')
plt.ylabel('oBSR')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting Weekly oBSR
plt.figure(figsize=(15, 5))
plt.plot(weekly_data['week'], weekly_data['oBSR'], marker='^', color='orange', label='Weekly oBSR')
plt.title('Weekly Overall Booking Success Rate (oBSR)')
plt.xlabel('Week')
plt.ylabel('oBSR')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```
